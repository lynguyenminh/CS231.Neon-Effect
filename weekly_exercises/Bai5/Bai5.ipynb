# Bài tập 5 - Phát hiện đường thẳng trên tập điểm

Bước 1: Generate tập điểm (n điểm) biết rằng 70% n thuộc 1 đường thẳng, 30% n không thuộc đường thẳng đó.

Bước 2: Chọn ngẫu nhiên 1 cặp điểm (A, B)

Bước 3: Tính phương trình tọa độ đường thẳng (d_AB)

Bước 4: Đếm số lượng điểm thuộc  (d_AB): m điểm thuộc (d_AB)

Bước 5: Nếu m >= 0.7*n thì vẽ đường thẳng lên biểu đồ và dừng

Ngược lại: quay lại Bước 2.

Input: Một mảng bất kì

Output: Đường thẳng đi qua các điểm



```python
import numpy as np
import matplotlib.pyplot as plt
import random 
```

```python
def check(y, x, a, b):
      return abs(a*x + b - y) < 0.00001
```

```python
def sol():

  #1. Khai báo số lượng phần tử trong mảng random
  Max = 1000
  Number = 100

  res = []

  #pt f(x) = 5x + 25
  #2. Thực hiện khởi tạo mảng bằng random
  x = np.array(random.sample(range(Max), Number))
  y = np.hstack(((5*x[:int(0.7*Number)] + 25), (random.sample(range(Max), int(Number * 0.3)))))
  x = x.reshape(-1, 1)
  y = y.reshape(-1, 1)
  df = np.hstack((x, y))

  #3. Thực hiện vòng lặp để duyệt tìm đường thẳng 
  while True:
    #4 chọn ra 2 điểm ngẫu nhiên
      count = 0
      temp = 2
      choices = np.random.choice(x.shape[0], temp, replace=False)
      print(choices)
      arr = df[choices]
      print(arr)
      ### BEGIN SOLUTION
      if arr[0,1] != arr[0, 0]:
          a = (arr[0,1] - arr[1, 1])/(arr[0, 0] - arr[1, 0])
          b = arr[0, 1] - a*arr[0, 0]
      else :
          continue
      for i, item in enumerate(df):
          count+= check(item[1], item[0], a, b)
          if check(item[1], item[0], a, b) == True:
            res.append([item[0], item[1]])
      ### END SOLUTION
      #5. vẽ đồ thị
      if count / df.shape[0] >= 0.7:
          plt.scatter(df[:, 0], df[:, 1])
          plt.plot([0, Max], [b, Max*a +b], 'r')
          plt.show()
          break

  return 

```

```python
sol()
```
