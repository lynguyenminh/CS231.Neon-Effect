```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Cover(img1_path, img2_path):
   #  Đọc 2 ảnh bằng hàm cv2.imread
  img1 = cv2.imread(img1_path)
  img2 = cv2.imread(img2_path)
  # Lấy thông tin chiều dài và rộng của ảnh
  h, w = img1.shape[:2]
  # scale 2 ảnh về cùng một kích thước
  img2 = cv2.resize(img2, [w, h])
  # Thực hiện duyệt qua từng pixel ảnh thực hiện phép toán
  for D in range(w - 1, 0, -1):
  ### BEGIN SOLUTION

  
    img1 = img1[:, :D]
    img2 = img2[:, :w-D]
    view = np.concatenate((img1, img2), axis = 1)

    plt.imshow(cv2.cvtColor(view, cv2.COLOR_BGR2RGB))
    plt.show()
  ### END SOLUTION
  return view
```

```python
def Push_Down(img1_path, img2_path):
   #  Đọc 2 ảnh bằng hàm cv2.imread
  img1 = cv2.imread(img1_path)
  img2 = cv2.imread(img2_path)
  # Lấy thông tin chiều dài và rộng của ảnh
  h, w = img1.shape[:2]
  # scale 2 ảnh về cùng một kích thước
  img2 = cv2.resize(img2, [w, h])
  # Tạo một ảnh đầu ra tạm thời
  view_pushdown = img1.copy()
  # Thực hiện duyệt qua từng pixel ảnh thực hiện phép toán
  for d in range(0, img1.shape[0]):
    ### BEGIN SOLUTION
    view_pushdown[0:d,:,:] = img1[img1.shape[0] - d :, :, :]
    view_pushdown[d:,:,:] = img2[0:img2.shape[0] - d,:,:]

    plt.imshow(cv2.cvtColor(view_pushdown, cv2.COLOR_BGR2RGB))
    plt.show()

    ### END SOLUTION
  return view_pushdown
```

```python
def Push_Left(img1_path, img2_path):
  #  Đọc 2 ảnh bằng hàm cv2.imread
  img1 = cv2.imread(img1_path)
  img2 = cv2.imread(img2_path)
  # Lấy thông tin chiều dài và rộng của ảnh
  h, w = img1.shape[:2]
  # scale 2 ảnh về cùng một kích thước
  img2 = cv2.resize(img2, [w, h])
  # Tạo một ảnh đầu ra tạm thời
  view_pushleft = img1.copy()
  # Thực hiện duyệt qua từng pixel ảnh thực hiện phép toán
  for d in range(0, img1.shape[1]):
    ### BEGIN SOLUTION

    view_pushleft[:, 0:d, :] = img1[:,img1.shape[1] - d : ,:]
    view_pushleft[:, d: ,:] = img2[:, 0:img2.shape[1] - d, :]
    
    plt.imshow(cv2.cvtColor(view_pushleft, cv2.COLOR_BGR2RGB))
    plt.show()
    ### END SOLUTION

  return view_pushleft
```

```python
def Push_up(img1_path, img2_path):
  #  Đọc 2 ảnh bằng hàm cv2.imread
  img1 = cv2.imread(img1_path)
  img2 = cv2.imread(img2_path)
  # Lấy thông tin chiều dài và rộng của ảnh
  h, w = img1.shape[:2]
  # scale 2 ảnh về cùng một kích thước
  img2 = cv2.resize(img2, [w, h])
  # Tạo một ảnh đầu ra tạm thời
  # Thực hiện duyệt qua từng pixel ảnh thực hiện phép toán
  for d in range(0, h):
    ### BEGIN SOLUTION
    img_1 = img1[d:]
    img_2 = img2[0:d]
    view_pushup = np.concatenate((img_1, img_2), axis = 0)
    plt.imshow(cv2.cvtColor(view_pushup, cv2.COLOR_BGR2RGB))
    plt.show()
    ### END SOLUTION

  return view_pushup
```

```python
def Uncover(img1_path, img2_path):
  #  Đọc 2 ảnh bằng hàm cv2.imread
  img1 = cv2.imread(img1_path)
  img2 = cv2.imread(img2_path)
  # Lấy thông tin chiều dài và rộng của ảnh
  h, w = img1.shape[:2]
  # scale 2 ảnh về cùng một kích thước
  img2 = cv2.resize(img2, [w, h])
  # Thực hiện duyệt qua từng pixel ảnh thực hiện phép toán
  for d in range(0, w):
    ### BEGIN SOLUTION
    img_1 = img1[:, d:]
    img_2 = img2[:, w - d:]
    view_uncover = np.concatenate((img_1, img_2), axis = 1)
    plt.imshow(cv2.cvtColor(view_uncover, cv2.COLOR_BGR2RGB))
    plt.show()
    ### END SOLUTION

  return view_uncover
```

```python
def sol(img1_path, img2_path):

 
  ### BEGIN SOLUTION
   #Thực hiện gọi các hàm tương ứng cho từng hiệu ứng
  cover = Cover(img1_path, img2_path)
  pushdown = Push_Down(img1_path, img2_path)
  pushup = Push_up(img1_path, img2_path)
  pushleft = Push_Left(img1_path, img2_path)
  uncover = Uncover(img1_path, img2_path)


  ### END SOLUTION
  return cover,pushup, pushdown, pushleft, uncover 
```
