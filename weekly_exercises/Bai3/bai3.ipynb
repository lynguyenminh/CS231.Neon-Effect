# Bài tập 2 - Cài đặt thuật toán Cross Correlation và Convolution





```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def sol(img): 
    # 1. Đọc ảnh tạo ma trận mask 
    img = cv2.imread(img, 0)
    mask = np.array([[-2, -2, -2],[0, 0, 0], [2, 2, 2]])

    #Thực Hiện cross corrletion

    # 2. lấy chiều cao, rộng của ảnh
    h,w = mask.shape
    # 3 Tạo ma trận để tính convolution

    Y = np.zeros((img.shape[0] - h + 1, img.shape[1] - w +1))

    #4. lặp qua các phần tử của ma trận mới và tính toán ra phần tử mới
    for i in range(Y.shape[0]):
      for j in range(Y.shape[1]):
    ### BEGIN SOLUTION
          Y[i, j] = (img[i : i +h, j : j+w] * mask).sum()

    ### END SOLUTION
    return Y   
```

```python
# thử nghiệm với ví dụ của đề bài
res =  sol('meo.png')

print(res.shape)

plt.imshow(res, cmap='gray')
plt.show()
```

```python
# cham diem voi public test
def evaluate(img1, img2):
  img1 = img1.astype(np.int16)
  img2 = img2.astype(np.int16)
  return np.mean(np.abs(img1 - img2))/255. < 0.01
  
testcase = cv2.imread('test_case.png', 0)
assert evaluate(res, testcase)
```
